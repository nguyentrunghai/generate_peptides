```python
import os
import PeptideBuilder
import Bio.PDB
```

```python
AA_LIST = "ACDEFGHIKLMNPQRSTVWY"
len(AA_LIST)
```

```python
for aa in AA_LIST:
    print(PeptideBuilder.Geometry.geometry(aa))
```

```python
def generate_aa(aa, pdb_path):
    structure = PeptideBuilder.make_extended_structure(aa)
    structure = PeptideBuilder.add_terminal_OXT(structure)
    out = Bio.PDB.PDBIO()
    out.set_structure(structure)
    out.save(pdb_path)
```

```python
generate_aa(AA_LIST, "test/20aa.pdb")
```

```python
generate_aa("AAA", "test/aaa.pdb")
```

# dipeptides

```python
for aa1 in AA_LIST:
    for aa2 in AA_LIST:
        aa = aa1+aa2
        pdb_path = os.path.join("peptide_build/dipeptides", aa+".pdb")
        print(pdb_path)
        generate_aa(aa, pdb_path)
```

# tripeptides

```python
for aa1 in AA_LIST:
    for aa2 in AA_LIST:
        for aa3 in AA_LIST:
            aa = aa1+aa2+aa3
            pdb_path = os.path.join("peptide_build/tripeptides", aa+".pdb")
            print(pdb_path)
            generate_aa(aa, pdb_path)
```

```python
for aa1 in AA_LIST:
    for aa2 in AA_LIST:
        for aa3 in AA_LIST:
            for aa4 in AA_LIST:
                aa = aa1+aa2+aa3+aa4
                pdb_path = os.path.join("peptide_build/tetrapeptides", aa+".pdb")
                generate_aa(aa, pdb_path)
```
