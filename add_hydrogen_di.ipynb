```python
import os
from pdbfixer import PDBFixer
from simtk.openmm.app import PDBFile
```

```python
AA_LIST = "ACDEFGHIKLMNPQRSTVWY"
```

```python
AA_LIST = "ACDEFGHIKLMNPQRSTVWY"

IN_DIR = "peptide_build/dipeptides"
OUT_DIR = "add_Hydrogens/dipeptides"
```

```python
for aa1 in AA_LIST:
    for aa2 in AA_LIST:
        pdb_inp = os.path.join(IN_DIR, aa1+aa2+".pdb")
        pdb_out = os.path.join(OUT_DIR, aa1+aa2+".pdb")
        print("Processing", pdb_inp)

        fixer = PDBFixer(pdb_inp)
        fixer.addMissingHydrogens(7.0)
        PDBFile.writeFile(fixer.topology, fixer.positions, open(pdb_out, 'w'))
```
